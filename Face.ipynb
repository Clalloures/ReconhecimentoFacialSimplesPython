## Parte 5: Destacando rosto

```python
import numpy as np
import cv2
# Haad Cascade XML
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

cap = cv2.VideoCapture(0)
while (True):
    # Captura frame-by-frame
    ret, frame = cap.read()
    # Converter a imagem em escala de cinza
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.5, minNeighbors=5, minSize=(30,30))
    
    for (x,y,w,h) in faces:
        roi_gray = gray[y:y+h, x:x+w]
        img_item_gray = 'my_image_gray.png'
        cv2.imwrite(img_item_gray, roi_gray)
        roi_color = frame[y:y+h, x:x+w]      
        img_item_color = 'my_image_color.png'
        cv2.imwrite(img_item_color, roi_color)
        
        color = (255,0,255) #BGR 0-255
        cv2.rectangle(frame, (x,y), (x+w, y+h), color, 2)
        
    # Mostra imagem
    cv2.imshow('frame', frame)
    # Como finalizar (apertar 'q')
    if cv2.waitKey(20) & 0xFF == ord('q'):
        break

# Fechar tudo
cap.release()
cv2.destroyAllWindows()
```

## Parte 2: Treinamento

O algoritmo Local Binary Pattern Histogram (LBPH) é uma solução simples para o problema de reconhecimento de face, que pode reconhecer a face frontal e a face lateral. No entanto, a taxa de reconhecimento do algoritmo LBPH nas condições de diversificação de iluminação, variação de expressão e deflexão de atitude é diminuída. Para resolver este problema, um algoritmo LBPH modificado baseado na mediana de cinza da vizinhança de pixel (MLBPH) é proposto. O valor cinza do pixel é substituído pelo valor mediano de seu valor de amostragem de vizinhança e, em seguida, o valor da característica é extraído pelos sub-blocos e o histograma estatístico é estabelecido para formar o dicionário de características MLBPH, que é usado para reconhecer o rosto humano identidade comparada com a imagem de teste. Os experimentos são realizados no banco de dados de faces padrão FERET e na criação de um novo banco de dados de faces, e os resultados mostram que o algoritmo MLBPH é superior ao algoritmo LBPH na taxa de reconhecimento.

```python
import cv2
import os
import numpy as np
from PIL import Image #Pillow : tabalha com imagens (trabalhar com diferentes formatos de imagens)
import pickle #mexer com arquivo


# Passar o diretorio para nosso dataset
image_dir = os.path.join('Face_Imagens/Dataset')

# Importar nosso xml
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')


# Sistema de reconhecimento: LBPHFaceRecognizer
# Existem várias opções no opencv, essa daqui é a mais utilizada
recognizer = cv2.face.LBPHFaceRecognizer_create()



current_id = 0
label_ids = {} #nome
y_labels = [] #id
x_train = [] #imagem

for root, dirs, files in os.walk(image_dir):
    for file in files:
        if file.endswith("png") or file.endswith("jpg"):
            path = os.path.join(root, file)
            label = os.path.basename(root)
            #print(label, path)
            if not label in label_ids:
                label_ids[label] = current_id
                current_id += 1
                print(label)
            id_ = label_ids[label]
            
            #print(label_ids)
            #y_labels.append(label) # some number
            #x_train.append(path) # verify this image, turn into a NUMPY arrray, GRAY
            pil_image = Image.open(path).convert("L") # grayscale
            size = (300, 300)
            final_image = pil_image.resize(size, Image.ANTIALIAS)
            image_array = np.array(final_image, "uint8")
            #print(image_array)
            faces = face_cascade.detectMultiScale(image_array, scaleFactor=1.5, minNeighbors=5)

            for (x,y,w,h) in faces:
                roi = image_array[y:y+h, x:x+w]
                x_train.append(roi)
                y_labels.append(id_)
                print(id_)


#print(y_labels)
#print(x_train)

with open("face-labels.pickle", 'wb') as f:
    pickle.dump(label_ids, f)

recognizer.train(x_train, np.array(y_labels))
recognizer.save("face-trainner.yml")
```

## Parte 3: Teste 

```python
import numpy as np
import cv2

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
recognizer = cv2.face.LBPHFaceRecognizer_create()
recognizer.read("face-trainner.yml")


labels = {"person_name": 1}

with open("face-labels.pickle", 'rb') as f:
    original_labels = pickle.load(f)
    labels = {v:k for k,v in original_labels.items()}

    
cap = cv2.VideoCapture(0)

while (True):
    # Captura frame-by-frame
    ret, frame = cap.read()
    # Converter a imagem em escala de cinza
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.5, minNeighbors=5)
    
    for (x,y,w,h) in faces:
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = frame[y:y+h, x:x+w]
        
        # Como reconhecer?
        id_, conf = recognizer.predict(roi_gray)
        if conf>=45:
            print(id_)
            print(labels[id_])
            font = cv2.FONT_HERSHEY_SIMPLEX
            name = labels[id_]
            color = (255,255,255)
            stroke = 2
            cv2.putText(frame, name, (x-5,y), font, 1, color, stroke, cv2.LINE_AA)
            
            color = (255,0,255) #BGR 0-255
            cv2.rectangle(frame, (x,y), (x+w, y+h), color, 2)
        
    # Mostra imagem
    cv2.imshow('frame', frame)
    # Como finalizar (apertar 'q')
    if cv2.waitKey(20) & 0xFF == ord('q'):
        break

# Fechar tudo
cap.release()
cv2.destroyAllWindows()
```
